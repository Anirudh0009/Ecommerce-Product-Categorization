## Data Exploration and Preparation:

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
df1 = pd.read_csv("test_data.csv")  

# Display basic information about the dataset
print("Dataset information:")
print(df1.info())

# Display summary statistics of numerical features
print("\nSummary statistics of numerical features:")
print(df1.describe())

# Check for missing values
print("\nMissing values:")
print(df1.isnull().sum())

# Check for duplicates
print("\nDuplicates:")
print(df1.duplicated().sum())

# Preprocess text data
# For example, remove special characters, lowercase text, handle missing values, etc.
# Here's a simple example for removing special characters and converting text to lowercase:
df1['description'] = df1['description'].str.replace('[^\w\s]', '').str.lower()

# Display a sample of preprocessed text data
print("\nSample of preprocessed text data:")
print(df1['description'].head())
```

```python
#Loading the train_product dataset 
train_df = pd.read_csv('train_product_data.csv')

# Display basic information about the dataset
print("Dataset information:")
print(train_df.info())

# Display summary statistics of numerical features
print("\nSummary statistics of numerical features:")
print(train_df.describe())

# Check for missing values
print("\nMissing values:")
print(train_df.isnull().sum())

# Check for duplicates
print("\nDuplicates:")
print(train_df.duplicated().sum())
```

```python
#Loading the second dataset 
df2 = pd.read_csv('test_results.csv')

# Display basic information about the dataset
print("Dataset information:")
print(df2.info())

# Display summary statistics of numerical features
print("\nSummary statistics of numerical features:")
print(df2.describe())

# Check for missing values
print("\nMissing values:")
print(df2.isnull().sum())

# Check for duplicates
print("\nDuplicates:")
print(df2.duplicated().sum())

```

```python
# Concatenate datasets vertically
df = pd.concat([df1, df2], axis=1)
```

```python
# Clean the data (e.g., remove special characters, lowercase text)
df['description'] = df['description'].str.replace('[^a-zA-Z0-9\s]', '').str.lower()

# Handle missing values (e.g., fill with a placeholder or impute)
df['description'].fillna("unknown", inplace=True)
```

## Fine Tuning

```python
#Hyperparameter Tuning

# Hyperparameter tuning for Logistic Regression
from sklearn.model_selection import GridSearchCV

lr_param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
lr_grid_search = GridSearchCV(lr_classifier, lr_param_grid, cv=5, scoring='accuracy')
lr_grid_search.fit(X_train, y_train)

print("Best Parameters for Logistic Regression:", lr_grid_search.best_params_)
```

```python
# Hyperparameter tuning for Random Forest
rf_param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
rf_grid_search = GridSearchCV(rf_classifier, rf_param_grid, cv=5, scoring='accuracy')
rf_grid_search.fit(X_train, y_train)

print("Best Parameters for Random Forest:", rf_grid_search.best_params_)
```
